```python
import pandas as pd
import numpy as np

df=pd.read_csv('fighters.csv',index_col=0).dropna().reset_index()
#%
df=df.drop('Id',axis=1)
#%%
df_B=df['B_Name']
df_R=df['R_Name']

df_all=df_B.append(df_R).drop_duplicates().reset_index().drop('index',axis=1)
#%%
df_all['Rating']=2000
df_all['Fights_Count']=0
#%%
df_all=df_all.rename(columns={0:'Name'})
#%%
df_all=df_all.set_index('Name')
#%%TESTY

#%%
df.insert(loc=3, column='B_Rating', value=0)   
df.insert(loc=4, column='B_Fights_Count', value=0)   
df.insert(loc=8,column='R_Rating',value=0)  
df.insert(loc=9, column='R_Fights_Count', value=0) 

#%%RATING
##################TUTAJ NIE WIEM JAKIEGO WZORU UZYC DO RATINGU WIEC NA RAZIE 
##################DODAJE 10 KIEDY ZAWONIK WYGRA I ODEJMUJE 10 GDY PRZEGERA
for key in range(df.shape[0]): 
    #Z tabeli z ratingiem wyciagamy rating poszczegolnych zawodnikow
    b_rating=df_all.at[df['B_Name'][key],'Rating']
    r_rating=df_all.at[df['R_Name'][key],'Rating']
    
    #Z tabeli z ratingiem wyciagamy ilosc walk poszczegolnych zawodnikow
    b_fights_count=df_all.at[df['B_Name'][key],'Fights_Count']
    r_fights_count=df_all.at[df['R_Name'][key],'Fights_Count']
    
    #Rating wrzucamy do głównej tabeli
    df['B_Rating'][key]=b_rating
    df['R_Rating'][key]=r_rating
    
    #Ilosc walk wrzucamy do glownej tabeli
    df['B_Fights_Count'][key]=b_fights_count
    df['R_Fights_Count'][key]=r_fights_count
    
    #Obliczamy rating po walce i wrzucamy go spowrotem do tabeli ratingowej
    #Wygrywa zawodnik niebieski
    if df['winner'][key]=='blue':
        df_all.at[df['B_Name'][key],'Rating']+=10
        df_all.at[df['R_Name'][key],'Rating']-=10
    #Wygrywa zawodnik czerwony
    if df['winner'][key]=='red':
        df_all.at[df['R_Name'][key],'Rating']+=10
        df_all.at[df['B_Name'][key],'Rating']-=10
        
    #Po walce dodajemy do tabeli +1 walke dla kazdego zawodnika
    df_all.at[df['B_Name'][key],'Fights_Count']+=1
    df_all.at[df['R_Name'][key],'Fights_Count']+=1
#Do ewentualnych zmian
   #Wygrywa zawodnik niebieski
    #if df1['winner'][key]=='blue':
     #   df_all.at[df1['B_Name'][key],'Rating']+=0.1*r_rating
      #  df_all.at[df1['R_Name'][key],'Rating']-=0.1*b_rating
    #Wygrywa zawodnik czerwony
    #if df1['winner'][key]=='red':
     #   df_all.at[df1['R_Name'][key],'Rating']+=0.1*b_rating
      #  df_all.at[df1['B_Name'][key],'Rating']-=0.1*r_rating


#%%Zmiana etykiet na wartosci liczbowe 
labels=df.iloc[:,-1:]
for key in range(labels.shape[0]):
    if labels['winner'][key]=='blue':
        labels['winner'][key]=0
    elif labels['winner'][key]=='red':
        labels['winner'][key]=1
    elif labels['winner'][key]=='draw':
        labels['winner'][key]=2
    elif labels['winner'][key]=='no contest':
        labels['winner'][key]=3
#%%
#%%Przygotowanie danych
data=((df-df.mean())/df.std()).dropna(axis=1)


#%%
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import linear
from tensorflow.keras.activations import sigmoid
from tensorflow.keras.activations import relu
from tensorflow.keras.activations import tanh
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dropout
labels=labels.to_numpy()
labels=to_categorical(labels,4)

#%%Budowanie modelu
#%%model 1
model=Sequential()
model.add(Dense(units=16,activation='relu',input_shape=(10,)))
model.add(Dropout(0.2))
model.add(Dense(units=8,activation='relu'))
model.add(Dense(units=4,activation='softmax'))

    
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])

history=model.fit(data,labels,batch_size=32,epochs=10,validation_split=0.3)
model_history=history.history
```
